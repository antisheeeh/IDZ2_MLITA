## 9. Исходный код программмы.

```python
i = -1
s = input()

def read():
    global i
    i += 1
    if i == len(s):
        return 0
    else:
        return s[i]

def error():
    print("Error")

def cbs(ch):
    if ch == '[':
        ch = read()
        ch = sqr(ch)
        if ch == ']':
            ch = read()
            ch = cbs(ch)
        else:
            error()
    elif ch == '(':
        ch = read()
        ch = cbs(ch)
        if ch == ')':
            ch = read()
            ch = cbs(ch)
        else:
            error()
    return ch

def sqr(ch):
    ch = var(ch)
    ch = var(ch)
    ch = cbs(ch)
    return ch

def var(ch):
    if ch == ')' or ch == ']':
        error()
        return ch
    
    if ch == '(':
        ch = read()
        ch = cbs(ch)
        if ch == ')':
            ch = read()
        else:
            error()
    elif ch == '[':
        ch = read()
        ch = sqr(ch)
        if ch == ']':
            ch = read()
        else:
            error()
    return ch

ch = read()
ch = cbs(ch)
```
